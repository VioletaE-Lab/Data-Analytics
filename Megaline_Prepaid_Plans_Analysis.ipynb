## Agregar datos por usuario


```python
# Calculate the number of calls made by each user per month.
calls_per_month = calls_df.groupby(["user_id","month"]).agg(
    calls_count = ("duration_rounded","count")
).reset_index()

display(calls_per_month.head(10))
```

```python
# Calculate the total minutes used by each user per month.
minutes_per_month = calls_df.groupby(["user_id","month"]).agg(
    minutes_used = ("duration_rounded","sum")
).reset_index()
display(minutes_per_month.sample(5))
```

```python
# Calculate the number of messages sent by each user per month.
sms_per_month = messages_df.groupby(["user_id","month"]).agg(
    sms_count = ("month","count")
).reset_index()
display(sms_per_month.sample(5))
```

```python
# Calculate the total GB used by each user per month.
internet_per_month = internet_df.groupby(["user_id","month"]).agg(
    gb_per_month = ("gb_used","sum")
).reset_index()
internet_per_month["gb_per_month_rounded"] = np.ceil(internet_per_month["gb_per_month"]).astype(int)
internet_per_month = internet_per_month.drop("gb_per_month", axis=1)
display(internet_per_month.head(5))
```

```python
user_plan_df = users_df[["user_id","plan"]]
user_plan_df.info()
```

```python
# Merge all the data frames into a single data frame called usage_per_month 

usage_per_month = minutes_per_month.merge(sms_per_month, on= ["user_id","month"], how= "outer")
usage_per_month = usage_per_month.merge(internet_per_month, on = ["user_id","month"], how="outer")
usage_per_month = usage_per_month.merge(user_plan_df, on = "user_id", how= "outer")
usage_per_month = usage_per_month.merge(calls_per_month, on = ["user_id","month"], how="outer")
#usage_per_month = usage_per_month.fillna(0)

# Drop rows where month is 0
usage_per_month = usage_per_month[usage_per_month["month"] != 0]


# Fill NaN values in numeric columns with 0
numeric_cols = ["minutes_used", "sms_count", "gb_per_month_rounded","calls_count","month"]
usage_per_month[numeric_cols] = usage_per_month[numeric_cols].fillna(0)


# Drop rows where month is 0
#usage_per_month = usage_per_month[usage_per_month["month"] != 0]

# Reemplazar numero de mes con nombre
# month_names = {1: "Enero", 2: "Febrero", 3: "Marzo", 4: "Abril",
#               5: "Mayo", 6: "Junio", 7: "Julio", 8: "Agosto",
#               9: "Septiembre", 10: "Octubre", 11: "Noviembre", 12: "Diciembre"}

#usage_per_month["month"] = usage_per_month["month"].replace(month_names)

usage_per_month["month"] = usage_per_month["month"].astype(int)
usage_per_month["minutes_used"] = usage_per_month["minutes_used"].astype(int)
usage_per_month["sms_count"] = usage_per_month["sms_count"].astype(int)
usage_per_month["gb_per_month_rounded"] = usage_per_month["gb_per_month_rounded"].astype(int)

#display(usage_per_month.head())
display(usage_per_month.sample(10))
```

```python
# Define the cost and included limits for each plan

surf_cost = 20 #$usd
surf_incld_min = 500 
surf_incld_sms = 50
surf_incld_gb = 15 
# If the limits of the package are exceeded:
surf_extra_min = 0.03 #cents
surf_extra_sms = 0.03 #cents
surf_extra_gb = 10  #usd

ultimate_cost = 70 #usd
ultimate_incld_min = 3000 
ultimate_incld_sms = 1000  
ultimate_incld_gb = 30
# If the limits of the package are exceeded:
ultimate_extra_min = 0.01 #cents
ultimate_extra_sms = 0.01 #cents
ultimate_extra_gb = 7  #usd
```

```python
# Define a function to calculate the monthly income for each user based on their plan and usage
surf_cost = 20 #$usd
surf_incld_min = 500 
surf_incld_sms = 50
surf_incld_gb = 15 
# If the limits of the package are exceeded:
surf_extra_min = 0.03 #cents
surf_extra_sms = 0.03 #cents
surf_extra_gb = 10  #usd

ultimate_cost = 70 #usd
ultimate_incld_min = 3000 
ultimate_incld_sms = 1000  
ultimate_incld_gb = 30
# If the limits of the package are exceeded:
ultimate_extra_min = 0.01 #cents
ultimate_extra_sms = 0.01 #cents
ultimate_extra_gb = 7  #usd

def calcular_ingreso(row):
    
    """
    Calculate the monthly revenue for a user based on their plan and usage.
    Parameters:
        row (pd.Series): A row from the usage DataFrame, containing 'plan',
                         'minutes_used', 'sms_count', 'gb_per_month_rounded', and 'month'.
    Returns:
        float: The total cost for the user in dollars.
"""
    plan = row['plan']
    minutes_used = row['minutes_used']
    sms_count = row['sms_count']
    gb_per_month_rounded = row['gb_per_month_rounded']
    month = row["month"]

    total_cost = 0.0

    if plan == "surf":
        total_cost += surf_cost
        if minutes_used > surf_incld_min:
            total_cost += (minutes_used - surf_incld_min) * surf_extra_min
        if sms_count > surf_incld_sms:
            total_cost += (sms_count - surf_incld_sms) * surf_extra_sms
        if gb_per_month_rounded > surf_incld_gb:
            total_cost += (gb_per_month_rounded - surf_incld_gb) * surf_extra_gb
    elif plan == "ultimate":
        total_cost += ultimate_cost
        if minutes_used > ultimate_incld_min:
            total_cost += (minutes_used - ultimate_incld_min) * ultimate_extra_min
        if sms_count > ultimate_incld_sms:
            total_cost += (sms_count - ultimate_incld_sms) * ultimate_extra_sms
        if gb_per_month_rounded > ultimate_incld_gb:
            total_cost += (gb_per_month_rounded - ultimate_incld_gb) * ultimate_extra_gb
            
    return total_cost

    # print("Original monthly usage DataFrame:")
    # display(usage_per_month)

    # Apply the function to calculate monthly revenue for each user
usage_per_month['monthly_income'] = usage_per_month.apply(calcular_ingreso, axis=1)

print("\nDataFrame with monthly revenue calculated:")
display(usage_per_month)
```

## Hypothesis Testing


### Hypothesis no. 1

The average income of Surf and Ultimate users differs.

Null hypothesis, H₀: The average income of Surf users equals that of Ultimate users.

Alternative hypothesis, H₁: The average income of Surf users differs from that of Ultimate users.

We'll use ttest_ind from SciPy, which is the independent two‑sample t‑test. This test is appropriate because we have two independent groups and want to compare their means. We will use an alpha level of 0.05: if the p-value is less than 0.05, we reject the null hypothesis

```python
# Hypothesis Testing 1: Surf vs Ultimate
surf_income 
ultimate_income

alpha = 0.05

t_stat, p_value = st.ttest_ind(surf_income,ultimate_income, equal_var=False)
print(f"T Stat: {t_stat:.4f}")
print(f"P Value: {p_value:.4f}")
print()

if p_value < alpha:
    print(f"P-value({p_value:.6f}) is less than alpha({alpha}).")
    print(f"We reject the null hypothesis.")
    print(f"The average income of Surf and Ultimate users is NOT equal; it IS different.")
else:
    print(f"P-value({p_value:.6f}) is greater than alpha ({alpha}).")
    print(f"We fail to reject the null hypothesis.")
    print(f"There is no evidence that the average income differs between the two plans.")
```

### Hypothesis no. 2

Null hypothesis, H₀: The average income of NY-NJ users is equal to that of other regions.

We will again use Student's t‑test with alpha = 0.05.
